```python
import os
import shutil
import natsort

file_names =  natsort.natsorted(os.listdir(path_img))
f = open(path)
data = json.load(f)

print("files len : ", len(file_names))
print("data json len : ", len(data["annotations"]))

maxlen = max(len(file_names), len(data["annotations"]))

img=0
for i in range(0, maxlen) : 
  id = data["annotations"][i]["id"]
  text = data["annotations"][i]["text"]
  
  name = file_names[img][0:-4] # 해당 순서의 이미지 이름.=이미지 번호
  if(int(name) > int(id)) : continue 
  elif(name==id) : 
    src = os.path.join(path_img, file_names[img])
    dst = text +"_"+ str(img+1) + ".png" #새로 저장할 이름
    shutil.copyfile(src, save_img + dst)
    # print("src : ",src)
    # print("img : ", img)
    # print("name: ", name)
    # print("i : ",i)
    # print("id : ", id)
    # print("text: ", text)
    img+=1
  if(i>50) : break
```
